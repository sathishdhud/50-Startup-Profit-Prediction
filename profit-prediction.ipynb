```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

df = pd.read_csv('/kaggle/input/profit/50_Startups (1).csv')
dataset = pd.DataFrame(df)
sns.histplot(data=dataset,bins=20,stat="count")

warnings.filterwarnings('ignore')
```

```python
f,ax = plt.subplots(2,2,figsize=(15,15))
sns.histplot(data=dataset,x="R&D Spend",bins=20,ax= ax[0][0],color="#56acdb")
sns.histplot(data=dataset,x="Administration",bins=20,ax=ax[0][1],color="#f28c2c")
sns.histplot(data=dataset,x="Marketing Spend",bins=20,ax=ax[1][0],color="#eb9494")
sns.histplot(data=dataset,x="Profit",bins=20,ax=ax[1][1],color="#63a86e")
f.tight_layout()
```

```python
#scatter plot
f,axs = plt.subplots(1,2, figsize=(12,5))
sns.scatterplot(data=dataset,ax=axs[0])
sns.barplot(data=dataset,ax=axs[1])
f.tight_layout()
```

```python
pd.plotting.scatter_matrix(dataset,c=list(map(lambda x:x%2,[i for i in range(50)])))
plt.figure(figsize=(24,24))
plt.show()
```

```python
from sklearn.model_selection import train_test_split

df = pd.DataFrame(dataset.iloc[:,:4])

x = df.iloc[:,:3]
y = df.iloc[:,3]

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)

#visualisation for Train dataset
sns.histplot(data=x_train)
sns.displot(data=x_train,kind="ecdf")
```

```python
#Visualisation for test dataset
sns.histplot(data=x_test)
sns.displot(data=x_test,kind="ecdf")
```
